## Fourier Transforms

* Consider a function defined at every point ${\bf x}$ in some volume.  Our main application will be the matter over-density field, $\delta({\bf x}) = \frac{\rho({\bf x}) - \bar{\rho}}{\bar{\rho}}$ We will define its Fourier Transform to be

<h3 align="center">
$\delta_{\bf k} = \int d^3x \delta({\bf x}) e^{-i {\bf k} \cdot {\bf x}} $   [note: units of volume!!].
</h3>
and the inverse is given by
<h3 align="center">
$\delta({\bf x}) = \int \frac{d^3k}{(2\pi)^3} \delta_{\bf k} e^{i {\bf k} \cdot {\bf x}} $   [unitless]
</h3>

* The Dirac delta function can be represented as
<h3 align="center">
$\delta^D({\bf k}) = \int \frac{d^3x}{(2\pi)^3} e^{i {\bf k} \cdot {\bf x}} $
</h3>
* [These notes](http://arxiv.org/pdf/0712.3028v2.pdf) contain expressions for Discrete Fourier Transforms.  Check out the [FFTW library](http://www.fftw.org/) when you need to compute a Fourier Transform numerically.

## Random Fields

## Properties of Gaussian Random Fields
* The one-point distribution: 
<h3 align="center">
$P(\delta({\bf x_1})) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-(\delta({\bf x_1})-\mu_1)^2/2\sigma_1^2}$.
</h3>
Here $\mu_1$ and $\sigma^2_1$ are the mean and variance of the field at point ${\bf x_1}$.


* The two-point distribution:  
<h3 align="center">
$P(\delta({\bf x_1}), \delta({\bf x_2})) = \frac{1}{2\pi|C|^{1/2}} e^{-\frac{1}{2}(\delta({\bf x_i})-\mu_i) C^{-1}_{ij} (\delta({\bf x_j})-\mu_j)}$
</h3>
Here C is the covariance matrix $C_{ij} = <(\delta({\bf x_i})-\mu_i)(\delta({\bf x_j})-\mu_j)>$, $C^{-1}$ is its inverse, and $|C|$ is its determinant.  The indices i and j run from 1 to 2.


* Now for the magic!  The joint probability for the Gaussian random field at $n$ points only depends on the two-point correlation function between the points, $C_{ij}$ (for i,j running from 1 to n).  Restricting ourselves to a field with mean 0 everywhere, we get:
<h3 align="center">
$P(\delta({\bf x_1}), ..., \delta({\bf x_n})) = \frac{1}{2\pi|C|^{1/2}} e^{-\frac{1}{2}(\delta({\bf x_i})) C^{-1}_{ij} (\delta({\bf x_j}))}$
</h3>


* In a homogeneous and isotropic universe, the two-point correlation function depends only on the magnitude of the separation vector between the points, not on ${\bf x_1}$ and ${\bf x_2}$ separately:
<h3 align="center">
$C_{ij} \equiv \xi(|{\bf x_i} - {\bf x_j}|)$
</h3>


* This implies that the Fourier modes are independent:
<h3 align="center">
$<\delta_{\bf k_1} \delta_{\bf k_2}> = \int d^3x_1 d^3x_2 <\delta({\bf x_1})\delta({\bf x_2})> e^{-i {\bf k_1} \cdot {\bf x_1}} e^{-i {\bf k_2} \cdot {\bf x_2}}$
</h3>

change variables to ${\bf r} = ({\bf x_1} + {\bf x_2})/2$ and ${\bf y} = {\bf x_2} - {\bf x_1}$:

<h3 align="center">
$<\delta_{\bf k_1} \delta_{\bf k_2}> = \int d^3r d^3y \xi(|{\bf y}|) e^{-i ({\bf k_1}+{\bf k_2}) \cdot {\bf r}} e^{-i ({\bf k_2}-{\bf k_1}) \cdot {\bf y}/2}$
</h3>

integrate over $d^3r$:

<h3 align="center">
$<\delta_{\bf k_1} \delta_{\bf k_2}> = (2\pi)^3 \delta^D({\bf k_1}+{\bf k_2}) \int d^3y \xi(|{\bf y}|) e^{-i {\bf k_2} \cdot {\bf y}} = (2\pi)^3 \delta^D({\bf k_1}+{\bf k_2}) P(|{\bf k_2}|)$
</h3>

like the correlation function, in an homogeneous and isotropic universe, the power spectrum depends only on the magnitude of the wavevector ${\bf k_2}$.

## Dynamics: gravity

An extremely simple and useful result from linear perturbation theory is that once the universe is deep in the matter-dominated (or dark-energy dominated) epoch, all modes of interest are growing at the same rate and phases remain unchanged:

<h3 align='center'>
$\delta_{\bf k}(a) = \delta_{\bf k}(a_i) \frac{D(a)}{D(a_i)}$
</h3>

Therefore the scale dependence of $P(k)$ remains unchanged, and its amplitude growths as $\propto D(a)^2$, where $D(a)$ is the linear growth rate.  Moreover, if $\delta_{\bf k}(a_i)$ is Gaussian, so is $\delta_{\bf k}(a)$ (as long as linear perturbation theory is valid).

## The Poisson Process

See Section 33 of [Peebles]((http://books.google.com/books/about/The_Large_scale_Structure_of_the_Univers.html?id=O_BPaHFtX1YC) and [MathWorld](http://mathworld.wolfram.com/PoissonDistribution.html) for more details.

* We will use the Poisson process to model how dark matter halos or galaxies ("tracers") are related to the underlying matter density field.  In this section, we will treat the underlying matter field as known/fixed, and only consider stochasticity in the positions of discrete tracers that sample the field.  Let those tracers have an expected number density $\bar{n}_t$.  Here's how it works:
    * Divide the space of interest (total volume $V$) up into a grid of tiny cells with volume $\delta V$, such that $\bar{n}_t \delta V << 1$.  In that case, each cell contains either 0 or 1 tracers.
    * The key assumption (vast simplification) is that each $\delta V_i$ is treated as independent.

**Question**: For a tracer with density 0.01 per $\delta V$, what is the probability of a configuration [1, 0, 0, 1, 0] for the first 5 volume elements?   

  * The probability that there are $k$ total tracers out of $N_{\rm cell} = V/\delta V$ cells is given by
    
<h3 align="center">
$P_{\lambda}(k) = \frac{\lambda^k e^{-\lambda}}{k!}$
</h3>
where $\lambda = \bar{n}_t*V$.  For large $\lambda$, the Poisson distribution approaches a Gaussian.

#### Poisson Process applied to large scale structure tracers (e.g. galaxies)
Now consider a model where the probability of a tracer in a particular cell depends on the local matter density:

<h3 align="center">
$p_t = F(\bar{\rho}_m (1 + \delta_m)) \delta V << 1$
</h3>

so the probability that the cell has 0 tracers is $(1-p_t)$.  As long as $\delta_m << 1$, we can Taylor expand $F$:

<h3 align="center">
$p_t \approx \bar{n}_t \delta V (1+b\delta_m)$
</h3>

Here $b$ is the ``bias'' of the tracer, and quantifies how sensitive the tracer density is to matter density fluctuations.  Let's simulate this for different values of $b$ and $n_t$.  We do this by splitting space into tiny cells and placing a galaxy in each cell independently with probability $p_t$.  The function below executes this.

```python
def Psample(k,A,xmin,xmax,Nt,b,dx=-1,dxbig=-1,ax=None,doplot=1,plotpoints=0,lbl=None,clr=None):
  """
  This function samples a sine wave with wavevector $k$ and amplitude $A$
  with between xmin and xmax with spacing dx.
  The mean number of tracers covering xmin to xmax is Nt,
  assuming xmax-xmin covers a unit number of oscillations of your wave.
  Tracers have bias b.
  This is a one-dimensional universe.
  dx should be chosen such that Nt*(1+A*b)*dx/(xmax-xmin) << 1,
  so that the probability for more than one galaxy in a cell of maximum
  p_t is still small.
  dx=-1 uses the code default for dx.
  dxbig is the setting for the histogram size.
  """
  if clr == None:
    clr = 'k'
  clr2 = clr + '--'
  if dx == -1:
    dx = 1.e-2*(xmax-xmin)/Nt/(1+A*b)
  x = np.arange(xmin,xmax+dx*0.5,dx)
  p_t = Nt/(xmax-xmin)*dx*(1.+b*A*np.sin(k*x))
  xx = np.where(np.random.random(len(x)) < p_t)[0]
  if dxbig == -1:
    dxbig = (xmax-xmin)/Nt*100
    dxbig = 0.01
  h, xtmp = np.histogram(x[xx],bins=np.arange(xmin,xmax,dxbig))
  ## you should really fit for the mean value as well.
  hdelta = (h-h.mean())/h.mean()
  xcen = 0.5*(xtmp[1:] + xtmp[:-1])
  dxcen = (xcen[1:]-xcen[:-1]).mean()

  hmodel = np.sin(k*xcen)
  ## ignore counts variation due to density fluctuation.
  #sig2 = Nt*dxcen/(xmax-xmin)*(1.+A*b*np.sin(k*xcen))
  sig2 = Nt*dxcen/(xmax-xmin)/(h.mean())**2
  s1 = (hdelta*hmodel/sig2).sum()
  s2 = (hmodel**2/sig2).sum()
  Aguess = s1/s2
  #print xcen
  #print hmodel
  #print hdelta
  #print s1, s2, Aguess, A*b
  if doplot == 1:
    if ax == None:
      f = plt.figure()
      ax = f.add_subplot(111)
    ax.plot(x,A*np.sin(k*x),clr2)
    if plotpoints==1:
      ax.plot(x[xx],A*np.sin(k*x[xx])*0.,'kx')
    ## also plot a histogram in fatter bins.
    ax.plot(xcen,hdelta,clr,label=lbl)
    #print 'produced this many tracers:',len(xx), len(xx)/Nt
    return ax, [Aguess, A*b, (s2.sum())**-0.5, (Aguess-A*b)/(s2.sum())**-0.5]

  else:  ## just return the best fit amplitude and best guess error (Gaussian approx)
    return [Aguess, A*b, (s2.sum())**-0.5, (Aguess-A*b)/(s2.sum())**-0.5]
Aval = 0.2
ax,stats=Psample(2*np.pi,Aval,0.,1.0,10000.,1.,lbl='b=1',clr='b')
print 'best fit %.3f +/- %.3f' % (stats[0], stats[2]),'expecting %.3f' % (stats[1])
ax,stats=Psample(2*np.pi,Aval,0.,1.0,10000.,4.,ax=ax,lbl='b=4',clr='g')
print 'best fit %.3f +/- %.3f' % (stats[0],stats[2]),'expecting %.3f' % (stats[1])
ax,stats=Psample(2*np.pi,Aval,0.,1.0,10000.,0.,ax=ax,lbl='b=0',clr='r')
print 'best fit %.3f +/- %.3f' % (stats[0],stats[2]),'expecting %.3f' % (stats[1])
plt.legend()
plt.axis([0,1,-1,1.2])

ax,stats=Psample(2*np.pi,Aval,0.,1.0,1000.,1.,lbl='N=1000',clr='b')
print 'best fit %.3f +/- %.3f' % (stats[0],stats[2]),'expecting %.3f' % (stats[1])
ax,stats=Psample(2*np.pi,Aval,0.,1.0,10000.,1.,ax=ax,lbl='N=10000',clr='g')
print 'best fit %.3f +/- %.3f' % (stats[0],stats[2]),'expecting %.3f' % (stats[1])
plt.legend()
plt.axis([0,1,-1,1.2])

## this test code demonstrates that the rms about the best fit A 
#is about what we predicted from our simple chi2 estimate, at least within a factor 2.
if 0==1:
  mtmp = 0
  stmp = 0
  ntrials=200
  for i in range(0,ntrials):
    stuff = Psample(2*np.pi,Aval,0.,1.0,10000.,4.,doplot=0)
    mtmp += stats[0]
    stmp += (stats[0]-stats[1])**2
  print 'mean, rms, expected mean, expected rms:',(mtmp/ntrials),(stmp/ntrials)**0.5, stats[1], stats[2]
```

## Fitting for the amplitude of single plane-wave fluctuation

* Bin the tracer counts in small spatial bins for size $dx$ across the domain $[x_{min}, x_{max}]$, so there are $n_{\rm bin} = (x_{max}-x_{min})/dx$ bins.  Our data vector is ${N_i}$, the list of counts in each cell.  The error on the counts in each bin is $\sim \sqrt{<N_i>}$.  We will neglect variations in the expected counts due to the fluctuation (i.e., $\delta << 1$), so the error on the counts in each bin is $\sqrt{N_t/n_{\rm bin}}$.  We will fit for the best fit amplitude for the sine wave with known frequency and phase.  Here is our expression for $\chi^2$:

<h3 align="center">
$\chi^2 = \sum_i \frac{(N_i - \hat{n}_t(1 + \hat{A} sin(k x_i))^2}{\sqrt{N_t / n_{\rm bin}}}$
</h3>

## Putting everything together.

* Define the overdensity field $\delta_t$ in terms of the number density field $n_t({\bf x}) = \bar{n}(1+\delta_t({\bf x}))$.

* Our final model for a discrete tracer overdensity field (like a set of galaxies) is a component proportional to the underlying matter fluctuations and a noise term, $\epsilon$, which accounts for the fact that we are sampling the continuous field with a finite number of tracers.
<h3 align="center">
$\delta_t({\bf k}) = b_t \delta_m({\bf k}) + \epsilon$
</h3>

* If we compute the power spectrum, we have a term proportional to the underlying matter power spectrum ($b^2_t P(k)$) and a shot noise term due to the Poisson sampling of the continous density field:
<h3 align="center">
$P_t({\bf k}) \equiv |\delta^2_t({\bf k})| = b^2_t P(k) + \bar{n}_t^{-1}$
</h3>
The signal that we want to measure if $b^2_t P(k)$, so we must remember to subtract off the shot noise contribution.

* The uncertainty in measuring the power $|\delta({\bf k})|^2$ in a single wavevector is (**Exercise 5**, show this!)
<h3 align="center">
$\sigma^2_P({\bf k}) = [P_t({\bf k})]^2 = [b^2_t P(k) + \bar{n}_t^{-1}]^2$
</h3>
The first term is called cosmic variance, and the second term is shot noise.

* Finally, the inverse signal-to-noise ratio for a single mode is
<h3 align="center">
$\frac{\sigma_{P({\bf k})}}{P({\bf k})} = \frac{b^2_t P(k) + \bar{n}_t^{-1}}{b^2_t P(k)}$ [single mode]
</h3>

* If we average over neighboring ${\bf k}$ to compute power spectrum bandpowers $P(k_i)$.  The uncertainty on the resulting bandpower shrinks by a factor $1/\sqrt{N_k}$, where $N_k = k_i^2 (\Delta k_i)(\Delta \mu_i) V_{\rm survey}/8\pi^2$ is the number of independent modes in the survey volume (think of density-of-states calculations in statistical physics).  

<h3 align="center">
$\frac{\sigma_{P(k_i)}}{P(k_i)} = \sqrt{\frac{8\pi^2}{k_i^2 (\Delta k_i)(\Delta \mu_i) V_{\rm survey}}}\frac{b^2_t P(k) + \bar{n}_t^{-1}}{b^2_t P(k)}$
</h3>

#### Exercise 5.

The equation above already tells you a lot about survey design.  Suppose your survey can take a total of $X = \bar{n}_t V_{\rm survey}$ spectra.  How should you trade off density and volume to get the smallest error on $P(k_i)$?

Solution: Choose the number density such that $\bar{n}_t b_t^2 P(k_i) = 1$.  In this case cosmic variance and shot noise contributions for every mode are equal.

* The Fisher matrix formalism provides a way to compute the expected error on a parameter that the power spectrum $P(k)$ is sensitive to (like the BAO scale, or the dark energy equation of state, ...).  The above equation for power spectrum errors provides the basis for those calculations.  See [Tegmark, Taylor, Heavens](http://adsabs.harvard.edu/abs/1997ApJ...480...22T) and [Seo and Eisenstein](http://adsabs.harvard.edu/abs/2003ApJ...598..720S) for more details.
