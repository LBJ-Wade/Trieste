## Probability Basics

* Normalization: $\int p(x) dx = 1$ and $p(x) \geq 0$.
* Moments: $<x^n> \equiv \int x^n p(x) dx$
* Joint Probability: $P(x,y)$
  * = $P(x) P(y)$ for independent events
* Conditional Probability: $P(x,y) = P(x|y) P(y) = P(y|x) P(x)$
* Bayes' Theorem (T = theoretical model with some parameters, D = data/observations): 
<h3 align="center">
$P(T | D) = \frac{P(D | T) P(T)}{P(D)}$
</h3>
$P(T)$ contains your "priors" (on parameters of T or T itself); you don't need to know $P(D)$ to compare different $T$ (it's just a normalization).

## Fourier Transforms

* Consider a function defined at every point ${\bf x}$ in some volume.  Our main application will be the matter over-density field, $\delta({\bf x}) = \frac{\rho({\bf x}) - \bar{\rho}}{\bar{\rho}}$ We will define its Fourier Transform to be

<h3 align="center">
$\delta_{\bf k} = \int d^3x \delta({\bf x}) e^{-i {\bf k} \cdot {\bf x}} $   [note: units of volume!!].
</h3>
and the inverse is given by
<h3 align="center">
$\delta({\bf x}) = \int \frac{d^3k}{(2\pi)^3} \delta_{\bf k} e^{i {\bf k} \cdot {\bf x}} $   [unitless]
</h3>

* The Dirac delta function can be represented as
<h3 align="center">
$\delta^D({\bf k}) = \int \frac{d^3x}{(2\pi)^3} e^{i {\bf k} \cdot {\bf x}} $
</h3>
* [These notes](http://arxiv.org/pdf/0712.3028v2.pdf) contain expressions for Discrete Fourier Transforms.  Check out the [FFTW library](http://www.fftw.org/) when you need to compute a Fourier Transform numerically.

## Random Fields
* A generic random field is daunting.  How are we going to learn about physics from an infinite list of random numbers, $\delta({\bf x})$?
* Our theory of cosmology does not predict the matter density at particular points in space, but rather its statistical properties.
* The initial conditions of our universe are extremeley close to Gaussian, which simplifies the math considerably.

## Properties of Gaussian Random Fields
* The one-point distribution: 
<h3 align="center">
$P(\delta({\bf x_1})) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-(\delta({\bf x_1})-\mu_1)^2/2\sigma_1^2}$.
</h3>
Here $\mu_1$ and $\sigma^2_1$ are the mean and variance of the field at point ${\bf x_1}$.


* The two-point distribution:  
<h3 align="center">
$P(\delta({\bf x_1}), \delta({\bf x_2})) = \frac{1}{2\pi|C|^{1/2}} e^{-\frac{1}{2}(\delta({\bf x_i})-\mu_i) C^{-1}_{ij} (\delta({\bf x_j})-\mu_j)}$
</h3>
Here C is the covariance matrix $C_{ij} = <(\delta({\bf x_i})-\mu_i)(\delta({\bf x_j})-\mu_j)>$, $C^{-1}$ is its inverse, and $|C|$ is its determinant.  The indices i and j run from 1 to 2.


* Now for the magic!  The joint probability for the Gaussian random field at $n$ points only depends on the two-point correlation function between the points, $C_{ij}$ (for i,j running from 1 to n).  Restricting ourselves to a field with mean 0 everywhere, we get:
<h3 align="center">
$P(\delta({\bf x_1}), ..., \delta({\bf x_n})) = \frac{1}{2\pi|C|^{1/2}} e^{-\frac{1}{2}(\delta({\bf x_i})) C^{-1}_{ij} (\delta({\bf x_j}))}$
</h3>


* In a homogeneous and isotropic universe, the two-point correlation function depends only on the magnitude of the separation vector between the points, not on ${\bf x_1}$ and ${\bf x_2}$ separately:
<h3 align="center">
$C_{ij} \equiv \xi(|{\bf x_i} - {\bf x_j}|)$
</h3>


* This implies that the Fourier modes are independent:
<h3 align="center">
$<\delta_{\bf k_1} \delta_{\bf k_2}> = \int d^3x_1 d^3x_2 <\delta({\bf x_1})\delta({\bf x_2})> e^{-i {\bf k_1} \cdot {\bf x_1}} e^{-i {\bf k_2} \cdot {\bf x_2}}$
</h3>

change variables to ${\bf r} = ({\bf x_1} + {\bf x_2})/2$ and ${\bf y} = {\bf x_2} - {\bf x_1}$:

<h3 align="center">
$<\delta_{\bf k_1} \delta_{\bf k_2}> = \int d^3r d^3y \xi(|{\bf y}|) e^{-i ({\bf k_1}+{\bf k_2}) \cdot {\bf r}} e^{-i ({\bf k_2}-{\bf k_1}) \cdot {\bf y}/2}$
</h3>

integrate over $d^3r$:

<h3 align="center">
$<\delta_{\bf k_1} \delta_{\bf k_2}> = (2\pi)^3 \delta^D({\bf k_1}+{\bf k_2}) \int d^3y \xi(|{\bf y}|) e^{-i {\bf k_2} \cdot {\bf y}} = (2\pi)^3 \delta^D({\bf k_1}+{\bf k_2}) P(|{\bf k_2}|)$
</h3>

like the correlation function, in an homogeneous and isotropic universe, the power spectrum depends only on the magnitude of the wavevector ${\bf k_2}$.

* Note that both the real and imaginary parts of $\delta({\bf k})$ are also Gaussian-distributed, independent of each other $(<Re[\delta({\bf k})] Im[\delta({\bf k})> = 0)$, and independent of every other Fourier mode ${\bf k_1} \neq {\bf k_2}$.  This makes it very easy to generate Gaussian random fields, and this is how [$N$-body simulation initial conditions](http://cosmo.nyu.edu/roman/2LPT/) are generated in practice.

## The Poisson Process

See Section 33 of [Peebles]((http://books.google.com/books/about/The_Large_scale_Structure_of_the_Univers.html?id=O_BPaHFtX1YC) and [MathWorld](http://mathworld.wolfram.com/PoissonDistribution.html) for more details.

* We will use the Poisson process to model how dark matter halos or galaxies ("tracers") are related to the underlying matter density field.  In this section, we will treat the underlying matter field as known/fixed, and only consider stochasticity in the positions of discrete tracers that sample the field.  Let those tracers have an expected number density $\bar{n}_t$.  Here's how it works:
    * Divide the space of interest (total volume $V$) up into a grid of tiny cells with volume $\delta V$, such that $\bar{n}_t \delta V << 1$.  In that case, each cell contains either 0 or 1 tracers.  If we neglect underlying fluctuations in the matter density, then the probability that there are $k$ total tracers out of $N_{\rm cell} = V/\delta V$ cells is given by
    
<h3 align="center">
$P_{\lambda}(k) = \frac{\lambda^k e^{-\lambda}}{k!}$
</h3>
where $\lambda = \bar{n}_t*V$.
    

Exercise 2: Show that $<k> = \sum_{k=0}^{\infty} k P_\lambda(k) = \lambda$, $<k^2> = \lambda^2 + \lambda$ and therefore $<(k-\bar{k})^2> = \lambda$.

Exercise 3: If you want to know the mean density of tracers to 1% and you can measure volume with negligible uncertainty, how many tracers must you count?  Answer: 10000.

Now consider a model where the probability of a tracer in a particular cell depends on the local matter density:

<h3 align="center">
$p_t = F(\bar{\rho}_m (1 + \delta_m)) \delta V << 1$
</h3>

so the probability that the cell has 0 tracers is $(1-p_t)$.  As long as $\delta_m << 1$, we can Taylor expand $F$:

<h3 align="center">
$p_t \approx \bar{n}_t \delta V (1+b\delta_m)$
</h3>

Here $b$ is the ``bias'' of the tracer, and quantifies how sensitive the tracer density is to matter density fluctuations.  Let's simulate this for different values of $b$ and $n_t$.

## To specify the expansion history of the universe in an FRW cosmology, we only need to specify the physical energy densities $\omega_x = \Omega_x h^2$ of all components today, and how they evolve with $a$.  We set a = 1 today, so $\sum \Omega_i = 1$ and $\sum \omega_i = h^2$, with the expansion rate at $a=1$ given by $H_0 = 10^5h$ m s$^{-1}$ Mpc$^{-1}$.

<h3 align="center">
$\left[\frac{1}{a} \frac{da}{dt}\right]^2 = H_0^2 [\Omega_{\gamma}a^{-4} + \Omega_{\nu}(a) + \Omega_m a^{-3} + \Omega_K a^{-2} + \Omega_{DE}(a)]$
</h3>
<h3 align="center">
$\left[\frac{1}{a} \frac{da}{dt}\right]^2 = \left(\frac{1}{9.785 \, {\rm Gyr}}\right)^2 [\omega_{\gamma}a^{-4} + \omega_{\nu}(a) + \omega_m a^{-3} + \omega_K a^{-2} + \omega_{DE}(a)]$
</h3>

For the energy density for thermal relic cosmological neutrinos $\Omega_{\nu}(a)$ see 3.3 of the [WMAP7 Cosmological Interpretation paper](http://adsabs.harvard.edu/abs/2011ApJS..192...18K), and for a detailed review, [Lesgourgues & Pastor 2006](http://arxiv.org/abs/astro-ph/0603494).

For flat $\Lambda$CDM with massless neutrinos, we have
<h3 align="center">
$\left[\frac{1}{a} \frac{da}{dt}\right]^2 = \left(\frac{1}{9.785 \, {\rm Gyr}}\right)^2 [\omega_{\gamma+\nu} a^{-4} + \omega_m a^{-3} + \omega_{\Lambda}]$
</h3>
and where $\omega_{\gamma+\nu} = 4.177 \times 10^{-5}$ is known precisely from measurements of the temperature of the cosmic microwave background (CMB).

```python
### set up a cosmology assuming a flat LambdaCDM cosmology and assuming massless neutrinos.

## define Omega_matter (omx), physical baryon density omega_baryon = Omega_baryon h^2 (obh2x), hubble constant little h (hx), 
## the spectral index (nsx), sigma8 (sig8x), physical cold dark matter density omega_cdm = Omega_cdm h^2 (och2x)
omx = 0.292
obh2x= 0.022
hx = 0.69
nsx=0.965
aeff=0.6452 ## output of the simulation box.
sig8x = 0.82
och2x = omx*hx**2 - obh2x
if 0==1:  ## computes mass function.
  ccPB=cosmo.cosmo(och2=och2x,obh2=obh2x,h=hx,forceflat=1,\
  pkfname="RunPB_pk_a0p6452.dat",zpk=1./0.6452-1.,\
  MFopt=1,mffname="PBMFtheory.dat")
else:  ## reads in mass fxn from file.
  ccPB=cosmo.cosmo(och2=och2x,obh2=obh2x,h=hx,forceflat=1,\
  pkfname="RunPB_pk_a0p6452.dat",zpk=1./0.6452-1.,\
  MFopt=2,mffname="PBMFtheory.dat")
    
print ccPB
print 'The critical density of the universe is ',cosmo.rhocrit, 'in units of (h^-1 Msun)/(h^-1 Mpc)^3'
rcrit = cosmo.rhocrit*((ccPB.h)**-1*cosmo.Msun)/((ccPB.h)**-1*cosmo.Mpc)**3
print 'in kg/m^3 that is:',rcrit
mH = 1.673534e-27 #https://www.cfa.harvard.edu/~dfabricant/huchra/ay145/constants.html
print 'which equals %.1f' % (rcrit/mH), 'hydrogen atoms per cubic meter.'

fs8x = ccPB.dDdlnaLCDM(a=aeff)*sig8x
print '\nThe growth rate f*sigma8 at z = %.3f was %.3f' % (1/aeff-1,fs8x)
```

```python
def rhorad(cc,a):
  return rcrit*(cc.og + cc.onu(alist))*alist**-4.

def rhom(cc,a):
  return rcrit*(cc.ocb)*alist**-3.

def rhoLambda(cc,a):
  return rcrit*(ccPB.oDE)+0.*alist

def rhoK(cc,a):
  return rcrit*np.fabs(-0.0020)*alist**-2  ## Planck + BOSS DR11 results finds Omega_k = -0.002 \pm 0.0047

ff=18
alist = 10**(np.arange(-7,3.0005,0.001))
print alist.min(), alist.max(), len(alist)
f=plt.figure(figsize=[10,5])
ax0=f.add_subplot(121)
ax0.loglog(alist,ccPB.Hofz(alist),'k')
ax0.set_xlabel('a',fontsize=ff)
ax0.set_ylabel('$H(a)$ [km/s/Mpc]',fontsize=ff)
ax1=f.add_subplot(122)
ax1.loglog(alist,rhorad(ccPB,alist),label='photons + neutrinos')
ax1.loglog(alist,rhom(ccPB,alist),label='cdm + baryons')
ax1.loglog(alist,rhoLambda(ccPB,alist),label='$\Lambda$')
ax1.loglog(alist,rhoK(ccPB,alist),'c--',label='curvature (current limit)')
plt.legend(loc=1)
ax1.set_xlabel('a',fontsize=ff)
ax1.set_ylabel(r'$|\rho_i(a)|$ [kg/m$^3$]',fontsize=ff)
#ax1.axis([0.9,1.1,rcrit*0.01,rcrit])
plt.tight_layout()
print ccPB.og + ccPB.onu(1.), ccPB.ocb, ccPB.oDE
```

```python
## rescale to physical units: k in Mpc^-1 and P(k) in Mpc^3.

k = ccPB.linPk.k*ccPB.h
pk = ccPB.linPk.pk*(ccPB.h)**-3

f=plt.figure(figsize=[11,11])
xx = np.where(k < 0.007)
plt.loglog(k, pk)
print 'asymptotic power law behavior and low/high k:',ccPB.linPk.nlow, ccPB.linPk.nhigh
plt.text(1.5e-4,5e3,r'Primordial Perturbations',fontsize=20)
#plt.text(1.5e-4,5e3,r'Outside the horizon during radiation',fontsize=20)
plt.text(1.5e-4,2.5e3,r'$P(k) \propto k^{n_s\approx 1}$',fontsize=20)
plt.axis([1.e-4,1,50,10e4])
plt.text(1.5e-4,5e3,r'Primordial Perturbations',fontsize=20)
plt.text(7e-4,6e4,r'~ Horizon Size at Matter-Radiation Equality',fontsize=20)
xmax = np.where(pk == pk.max())
plt.loglog(k[xmax],pk[xmax],'x',markersize=30,markeredgewidth = 3,color='k')
plt.xlabel(r'$k$ [Mpc$^{-1}$]',fontsize=20)
plt.ylabel(r'Linear matter power spectrum [Mpc$^{3}$]',fontsize=20)
plt.text(1.5e-2,5e3,r'Baryon Acoustic Oscillations',fontsize=20)
```
